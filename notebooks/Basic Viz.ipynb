```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
URL_NEW = 'https://covid.ourworldindata.org/data/jhu/new_cases.csv'
df = pd.read_csv(URL_NEW)

data = df.copy(deep=True)
```

```python
data.head()
```

```python
COUNTRY = 'India'           # Country/Region Name (example: INDIA / FRANCE / DENMARK)
TYPE = 'CONFIRMED'          # CONFIRMED / DEATHS / RECOVERED
SCALE = 'LOG'               # PERCENT / LOG
```

```python
country_new = data[COUNTRY.title()]

if country_new.isnull()[0]:
    country_new[0] = 0
    
country_new = country_new.ffill().copy()

country_new = country_new.where(country_new>=0, -country_new).copy()

## Check if log scale is selected
if SCALE=='+':
    ### log 0 is infinity. Fake one person 
    country_new = np.log(country_new+np.ones(len(country_new)))
```

```python
fig = go.Figure()

fig.add_trace(go.Histogram(x=data.date, 
                         y=country_new, marker_color='royalblue',
                         histfunc="avg", xbins_size="D1",
                         name=f'{TYPE} Cases'))

fig.add_trace(go.Scatter(x=data.date, 
                         y=country_new.rolling(7).mean(), mode='lines', 
                         line=dict(color='firebrick', width=2),
                         name='Average'))

fig.update_layout(bargap=0.2)

fig.show()
```
